```python
#Life Expectancy and GDP
#Introduction
#This is a portfolio project for codecademy on life expectancy and GDP. The data is from the World Health Organization and the World Bank. The goal of this project is to try and identify the relationship between life expectancy and GDP in six countries.

#Questions to answer:
#1). How are life expectancy and gdp correlated?
#2). How has life expectancy and gdp changed throughout the years in each country?
```

```python
# Importing Modules

import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Loading Data
gdp_life_proj = pd.read_csv('all_data.csv')
print(gdp_life_proj.head())
```

```python
# Exploratory Analysis
# Want to explore the unique values of countries and the different years used, while seeing the max and min of the Life Expectancy and GDP data
gdp_life_proj.Country.unique()
```

```python
gdp_life_proj.Year.unique()
```

```python
# Will have to rename the column name to make it easier to call when needed in the future, before calling the min and max values for life expectancy
gdp_life_proj.rename(columns = {'Life expectancy at birth (years)': 'Life_Exp'}, inplace=True)
```

```python
print(gdp_life_proj.Life_Exp.min(), gdp_life_proj.Life_Exp.max())
```

```python
print(gdp_life_proj.GDP.min(), gdp_life_proj.GDP.max())
```

```python
# Findings
# There are six countries: Chile, China, Germany, Mexico, United States of America, and Zimbabwe.
# The year spans from 2000 through 2015
# The Life Expectancy from birth spans from 44.3 at the miniumum through 81.0 years maximum
# The GDP in US dollars range from a minimum of 4 trillion to 18 trillion
```

```python
# Data Analysis
# To answer question #1 of how life expectancy is correlated with GDP?
```

```python
ax1 = sns.scatterplot(x = 'Life_Exp', y = 'GDP', hue = 'Year', data = gdp_life_proj)
plt.title('GDP vs Life Expectancy by Year')
ax1.set_xlabel('Life Expectancy')
ax1.set_ylabel('GDP (In 10s of Trillions)')
```

```python
# Figure 1: Is colorized by year, which shows that life expectancy and GDP are increasing over the years
```

```python
# Figure 2: GDP vs Life Expectancy shown colorized by country. Which shows that as GDP increases, so does life expectancy. 
# This is clearly true for China and the USA as their slopes are the steepest
# Only the exception of Zimbabwe which shows that while life expectancy is increasing, but not GDP is not changing. Due to possible infrastructure and maybe medicinal advancements
```

```python
ax1 = sns.scatterplot(x = 'Life_Exp', y = 'GDP', hue = 'Country', data = gdp_life_proj)
plt.title('GDP vs Life Expectancy by Country')
ax1.set_xlabel('Life Expectancy')
ax1.set_ylabel('GDP (In 10s of Trillions)')
```

```python
# Building Six Plots with Twin Axis of Life Expectancy and GDP vs. Years, to compare each country individually
```
